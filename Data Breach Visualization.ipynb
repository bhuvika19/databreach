```python
import pandas as pd # import the pandas library. pd is a common shorthand for pandas
import numpy as np # import numpy for working with numbers
import plotly.figure_factory as ff
import plotly.express as px
import plotly.graph_objects as go
```

```python
df = pd.read_csv('irgp.csv')
df.head()
```

```python
custom_colorscale = [
    [0.0, 'rgb(0,0,51)'],
    [0.1, 'rgb(0,0,102)'],
    [0.2, 'rgb(51,102,153)'],
    [0.4, 'rgb(153,204,255)'],
    [0.6, 'rgb(204,229,255)'],
    [1.0, 'rgb(230,230,250)']
]

fig = go.Figure(data=go.Choropleth(
    locations=df['Abbreviation'], # Spatial coordinates
    z = df['Signed'].astype(float), # Data to be color-coded
    locationmode = 'USA-states', # set of locations match entries in `locations`
    colorscale = custom_colorscale,
    autocolorscale=False,
    reversescale=False,
    text=df['State'],
    marker_line_color='white',
    colorbar_title = "Years",
))

fig.update_layout(
    title_text = 'Signing of Security Breach Notification Laws by State',
    geo_scope='usa', # limite map scope to USA
     width=1000, # Increase plot width
    height=800
)

fig.show()
```
